```python
import tkinter as tk

def save_data():
    name = name_entry.get()
    email = email_entry.get()
    password = password_entry.get()
    with open("users.txt", "a") as file:
        file.write(f"Name: {name}, Email: {email}, Password: {password}\n")
    name_entry.delete(0, tk.END)
    email_entry.delete(0, tk.END)
    password_entry.delete(0, tk.END)
    message_label.config(text="Successfully registered!")

root = tk.Tk()
root.title("Registration Page")

name_label = tk.Label(root, text="Name:")
name_label.grid(row=0, column=0)

name_entry = tk.Entry(root)
name_entry.grid(row=0, column=1)

email_label = tk.Label(root, text="Email:")
email_label.grid(row=1, column=0)

email_entry = tk.Entry(root)
email_entry.grid(row=1, column=1)

password_label = tk.Label(root, text="Password:")
password_label.grid(row=2, column=0)

password_entry = tk.Entry(root, show='*')
password_entry.grid(row=2, column=1)

register_button = tk.Button(root, text="Register", command=save_data)
register_button.grid(row=3, column=1)

message_label = tk.Label(root)
message_label.grid(row=4, column=1)

root.mainloop()

```
